```python
import yfinance as yf

# Fetch stock data
df = yf.download(
    "AAPL",
    start="2024-01-01",
    end="2025-01-01"
)

# Show first rows
print(df.head())
```

```python
df.to_csv("apple_ohlc_data.csv")

print("CSV Saved Successfully")
```

```python
df = yf.download(
    "AAPL",
    period="5d",
    interval="15m"
)

print(df.head())
```

```python
import matplotlib.pyplot as plt

df['Close'].plot(figsize=(12,5))

plt.title("AAPL Closing Price")
plt.show()
```
