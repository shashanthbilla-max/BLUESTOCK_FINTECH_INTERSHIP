```python
import pandas as pd
```

```python
df=pd.read_csv('D:/BULESTOCK_INTERSHIP/DAY 5 WORK/cleaned_nav_history.csv')
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df = df.sort_values(by='date')

print(df.head())
```

```python
fund_code = 125497

fund_df = df[df['amfi_code'] == fund_code]

print(fund_df.head())
```

```python
begin_nav = fund_df.iloc[0]['nav']

end_nav = fund_df.iloc[-1]['nav']

print("Beginning NAV:", begin_nav)
print("Ending NAV:", end_nav)
```

```python
start_date = fund_df.iloc[0]['date']

end_date = fund_df.iloc[-1]['date']

years = (end_date - start_date).days / 365.25

print("Years:", years)
```

```python
cagr = ((end_nav / begin_nav) ** (1 / years)) - 1

print(f"CAGR: {cagr * 100:.2f}%")
```
