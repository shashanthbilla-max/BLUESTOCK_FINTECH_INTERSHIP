```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("D:/BULESTOCK_INTERSHIP/DAY 3 WORK/financial_data.csv")
df
```

```python
plt.figure(figsize=(10,5))

sns.barplot(x='category',
            y='amount',
            data=df)

plt.title("Amount by Category")
plt.xlabel("Category")
plt.ylabel("Amount")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.hist(df['amount'],
         bins=20)

plt.title("Amount Distribution")
plt.xlabel("Amount")
plt.ylabel("Frequency")

plt.show()
```

```python
category_amount = df.groupby('category')['amount'].sum()

# Plot
plt.figure(figsize=(8,8))

plt.pie(category_amount,
        labels=category_amount.index,
        autopct='%1.1f%%')

plt.title("Category Wise Amount Distribution")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.scatterplot(x='transaction_id',
                y='amount',
                data=df)

plt.title("Transaction ID vs Amount")
plt.xlabel("Transaction ID")
plt.ylabel("Amount")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(x='payment_method',
            y='amount',
            data=df)

plt.title("Payment Method vs Amount")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(x='payment_method',
              data=df)

plt.title("Number of Transactions by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Transaction Count")

plt.xticks(rotation=45)

plt.show()
```
