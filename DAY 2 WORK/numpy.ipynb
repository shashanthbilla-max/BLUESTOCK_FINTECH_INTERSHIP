Indexing means:

Accessing individual elements or groups of elements from a NumPy array using positions (indexes).

```python
a = np.array([10,20,30,40,50])
```

```python
print(a[0])  # First element
print(a[1])  # Second element
print(a[2])  # Third element        
print(a[3])  # Fourth element
print(a[4])  # Fifth element
```

Negative Indexing

Negative indexing starts from end.

```python
print(a[-1])  # Last element
print(a[-2])  # Second to last element
print(a[-3])  # Third to last element
print(a[-4])  # Fourth to last element
print(a[-5])  # Fifth to last element
```

```python
a = np.array([1,2,3,4,5,6])
```

```python
print(a[1:5])
```

```python
print(a[::2])
```

```python
print(a[::-1])
```

merging:combine the two or more DATAFRAMES

```python
df1 = pd.DataFrame({
    "customerID": [1,2,3],
    "Name": ["Ravi","Sai","Teja"]
})
```

```python
df2 = pd.DataFrame({
    "customerID": [1,2,3],
    "Amount": [500,700,900]
})
```

```python
merged = pd.merge(df1, df2, on="customerID")

print(merged)
```
