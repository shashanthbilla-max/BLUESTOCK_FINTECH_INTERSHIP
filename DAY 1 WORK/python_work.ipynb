Dictionary (dict):A dictionary stores data in key-value pairs

Dictionary Features:
Stores data using keys,
Changeable,
Fast data access,
No duplicate keys

```python
d = {
    "abhi": 1,
    "sai": 2,
    "mahesh": 3,
    "teja": 4,
    "lucky": 5,
    "ravi": 5,
    "raju": 6
}

print(d)
print(type(d))

# Add
d["nikhil"] = 7
print(d)

# Update
d["abhi"] = 101
print(d)

# Multiple update
d.update({"bangairi": 7, "mohan": 8})
print(d)

# Add new key-value pair
d["gopi"] = 102
print(d)
```


Inheritance means:

One class gets properties and methods from another class.

```python
class Animal:

    def sound(self):
        print("Animal sound")

class Dog(Animal):

    def bark(self):
        print("Bark")
```

```python
d = Dog()

d.sound()
d.bark()
```
